**Below is a list of the column I am interested in, with annotations.**

artist_name
name of the recording artist

artist_name_clean
artist name all upper case, no spaces, with secondary artists ("featuring") removed.

track_name
name of the track, i.e. usually name of the song

first_entry
date of the first entry into the Billboard Hot 100

quarter, year, fiveyear, decade
transformations of first_entry to coarser time periods

harm_…
193 columns of chord change counts; the chord change is indicated in the column label (e.g. harm_M.2.M means major chord followed by another major chord 2 semitones up).


```python
# clean up data and keep only the columns I need(listed above)
# 'artist_name_clean, track_name, first_entry, ')
df2_m = df_m.iloc[:,(2,3,4,5, 41,42,43,44,45,46,47, 53,54,55,56,57,58,59)]
df2_LP = df2_m[df2_m.artist_name_clean == "LINKINPARK"]
df2_LP.dropna(how='any', inplace=True)
df2_LP.head()
```

```python
# Createa a new column with total chord changes

#df2_LP = pd.to_numeric(df2_LP.iloc[:,14:], errors='coerce').astype(float)
df2_LP['chord_changes'] = df2_LP.iloc[:,-14:].sum(axis=1, numeric_only=1)
df2_LP
```

```python
df2_LP.shape
```

```python
# Plot the bar chart using Bokeh packages

from bokeh.charts import Bar, output_file, show
from bokeh.charts.attributes import cat, color
import bokeh.palettes
from bokeh.charts.operations import blend
#from bokeh.charts.utils import df_from_json
from bokeh.plotting import *
from bokeh.models import Legend


output_notebook()

# if we need to get json/dict data converted to a dataframe
#df = df_from_json(data)

# filter by Linkin park sound tracks with at least one chord change and sort by total chord changes

df2_LP = df2_LP[df2_LP.chord_changes > 0]
df2_LP = df2_LP.sort_values(by = "chord_changes", ascending=False)


bar = Bar(df2_LP, 
          values=blend('harm_M.0.M', 'harm_M.1.M', 'harm_M.2.M', 'harm_M.3.M', 'harm_M.4.M', 'harm_M.5.M', 'harm_M.6.M',
                       'harm_M.0.m', 'harm_M.1.m', 'harm_M.2.m', 'harm_M.3.m', 'harm_M.4.m', 'harm_M.5.m', 'harm_M.6.m',
                       name='chord changes', labels_name='chord_change_to'),
          bar_width=0.8, 
          plot_width= 800,
          plot_height = 650,
          label=cat(columns='track_name', sort=False),
          stack=cat(columns='chord_change_to', sort=False),
          color=color(columns='chord_change_to',
                      palette=['AntiqueWhite','NavajoWhite', 'gold', 'orange', 'darkorange', 'coral', 'tomato',
                              'Azure', 'PaleTurquoise', 'lightskyblue', 'deepskyblue', 'dodgerblue','blue','blueviolet'],
                      sort=False),
          toolbar_location="right",
          legend = "top_right",
          xlabel = "Sound Track Name",
          title = 'Figure - Chord Changes per Sound Track(Linkin Park), Sorted by total chord changes',
          tooltips=[('chord changed to ','chord_change_to'), ('song', '@track_name')])

bar.title.text_font = 'Times'
bar.title.text_font_size = {'value': '15pt'}

#bar.background_fill_color = "white"

show(bar)
```
